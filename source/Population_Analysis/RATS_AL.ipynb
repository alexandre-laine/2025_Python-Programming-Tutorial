```python
import matplotlib.pyplot as plt
import scipy.stats as stt
import numpy as np
import os
```

```python
# parameters
cwd = os.getcwd()  # current working directory
ROOT = os.path.dirname(os.path.dirname(cwd))  # root directory

data_raw_file_name = "data_raw"  # folder with raw data
data_raw_path = os.path.join(ROOT, data_raw_file_name)  # path of this folder
```

```python
subject = "RAT24"
```

```python
fig, ax = plt.subplots(
    nrows=1, ncols=1,
    figsize=(30,5)
)

i = 0

for name in os.listdir(data_raw_path):

    if name[:len(subject)] == subject:

        data = np.load(os.path.join(data_raw_path, name), allow_pickle=True).item()
        
        ax.scatter(
            data["SpikeTiming"],
            np.ones_like(data["SpikeTiming"]) * i,
            marker="|"
        )

        i += 1

plt.show()
```
